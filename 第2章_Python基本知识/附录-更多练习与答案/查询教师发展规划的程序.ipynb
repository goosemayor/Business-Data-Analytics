# 练习：查询教师发展规划的程序  [挑战]

请用**循环语句**和**判断语句**来设计一个程序：

一个老师他输入自己的入职年限，程序就自动地按顺序返回学校给予这个老师的建议。

!!! Tips
    1. 写判断语句
    2. 在判断中嵌套循环
    3.  完成输入、输出功能



### 答案

第一步，写判断语句，判断一个老师输入的年限是1-2、3-4还是5-8年，编写如下代码，并测试通过。

```python
year = 1
if year >= 1 and year<=2:
    print('您的入职年限:', year, "属于教师发展的第一个阶段")
elif year >=3 and year <= 4:
    print('您的入职年限:', year, "属于教师发展的第二个阶段")
elif  year >=5 and year <= 8:
    print('您的入职年限:', year, "属于教师发展的第三个阶段")
    
```

第二步，写循环语句，我们在每个判断语句中，打印出4项学校给予这个老师的建议，例如前两年：

```python
suggestions = ["高校教师资格证",
              "初级职称认定",
              "强师工程",
              "申报校级课题与市级\"高青项目\""]
for i in range(4):
    print(i+1, suggestions[i])
```

第三步，我们把循环语句放入判断语句中

```python
if year >= 1 and year<=2:
    print('您的入职年限:', year, "属于教师发展的第一个阶段")
    suggestions = ["高校教师资格证",
                  "初级职称认定",
                  "强师工程",
                  "申报校级课题与市级\"高青项目\""]
    for i in range(4):
        print(i+1, suggestions[i])
```

依此类推

第四步，完善输入input语句

```python
year = input('请输入您的入职年限[1-8年]')
year = int(year)

if year >= 1 and year<=2:
    print('您的入职年限(年):', year, "属于教师发展的第一个阶段")
    suggestions = ["高校教师资格证",
                  "初级职称认定",
                  "强师工程",
                  "申报校级课题与市级\"高青项目\""]
    for i in range(4):
        print(i+1, suggestions[i])
        
elif year >=3 and year <= 4:
    print('您的入职年限(年):', year, "属于教师发展的第二个阶段")
    suggestions = ["中级职称认定",
                  "产学研践习",
                  "各类教师高级研修项目",
                  "投身教科研"]
    for i in range(4):
        print(i+1, suggestions[i])
        
elif  year >=5 and year <= 8:
    print('您的入职年限(年):', year, "属于教师发展的第三个阶段")
    suggestions = ["高级职称评审",
                  "产学研践习",
                  "各类教师高级研修项目",
                  "投身教科研"]
    for i in range(4):
        print(i+1, suggestions[i])
else:
    print("您输入的入职年限应该在[1-8年]")
```
