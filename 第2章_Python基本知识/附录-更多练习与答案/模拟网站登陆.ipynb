# 练习：模拟网站登陆  [容易]

我们通过以下程序，模拟网站登陆的情景。网站的登陆和登出需要账户名，我们先创建包含5个人名的列表，其中包含'admin'这个字符串。我们一般使用'admin'代表管理员。

1）一个用户登录网站后，网站显示问候语。例如，如果是用户名是'admin'，网站显示问候语：'hello，admin，would you like to see a status reports of users'；否则，打印"Hello Jason, thank you for logging in."

2）一个新用户，他想"Michael"作为他的用户名，如果他在网站已有的用户名列表中，则建议他使用换个名字，所以打印“This user name has been used, please try another user name”，提示用户输入新的用户名；如果这个名字不在已有的用户名列表中，就把这个名字加入已有的用户名列表中，并打印"Thanks for signing up!  Michael "

!!!Tips
    in 和 not in 表示元素在或不在某个集合中，例如 "a" in ["a", "b", "c"] 逻辑运算为True

## 答案

1）先创建包含5个人名的列表，其中包含'admin'这个字符串。

```python
name_list = ['Michael', 'Jane', 'Rose', 'Hans', 'admin']
```

判断登录名字是不是'admin'：

```python
name = 'Hans'
if name == 'admin':
    print('hello，admin，would you like to see a status reports of users')
else:
    print("Hello %s, thank you for logging in." %name)
```

2）我们使用判断语句，来判断一个用户在不在网站已有用户列表中：

- 如果在，就告诉用户换一个用户名
- 如果不在，就添加到用户列表中

```python
name = "Michael"
if name in name_list:
    print("This user name has been used, please try another user name：")
else:
    name_list.append(name) 
    print("Thanks for signing up! ", name)
```

如果用户继续输入的名字还在已有用户列表中呢？比如"Hans"。所以，每当用户输入名字，都要判断，直到用户输入一个网站已有用户列表没有的名字。

- 我们使用while循环语句来完成这个逻辑
- 当用户输入很多次之后，终于输入一个网站用户列表没有的名字的时候，我们需要跳出循环，终止程序

```python
name_list = ['Michael', 'Jane', 'Rose', 'Hans', 'admin']

name = "Michael"
while True:
    if name in name_list:
        name = input("This user name has been used, please try another user name：")
    else:
        name_list.append(name) 
        print("Thanks for signing up! ", name)
        break
```
