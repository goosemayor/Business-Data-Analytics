# 练习：改造乘积函数 [困难]

以下函数允许计算两个数的乘积，请稍加改造，变成可接收一个或多个数并计算乘积：

```python
def mul(x, y):
    return x * y

# Tips:
# 1）选择使用哪种参数类型：位置参数？默认参数？可变参数？关键字参数？
# 2）使用哪种语句组织逻辑：判断语句？循环语句？
# 3）不要忘记return

```

```python
# 使用以下代码对改造后的函数进行测试
print('mul(5) =', mul(5))
print('mul(5, 6) =', mul(5, 6))
print('mul(5, 6, 7) =', mul(5, 6, 7))
print('mul(5, 6, 7, 9) =', mul(5, 6, 7, 9))

if mul(5) != 5:
    print('测试失败!')
elif mul(5, 6) != 30:
    print('测试失败!')
elif mul(5, 6, 7) != 210:
    print('测试失败!')
elif mul(5, 6, 7, 9) != 1890:
    print('测试失败!')
else:
    try:
        mul()
        print('测试失败!')
    except TypeError:
        print('测试成功!')
```



## **答案**

第一步 选择可变参数

```python
def mul(*x):
```

第二步 选择循环语句

```python
def mul(*x):
    res = 1
    for i in x:
        res = res * i
    return res
```

第三步 测试

```python
print('mul(5) =', mul(5))
print('mul(5, 6) =', mul(5, 6))
print('mul(5, 6, 7) =', mul(5, 6, 7))
print('mul(5, 6, 7, 9) =', mul(5, 6, 7, 9))
if mul(5) != 5:
    print('测试失败!')
elif mul(5, 6) != 30:
    print('测试失败!')
elif mul(5, 6, 7) != 210:
    print('测试失败!')
elif mul(5, 6, 7, 9) != 1890:
    print('测试失败!')
else:
    try:
        mul()
        print('测试失败!')
    except TypeError:
        print('测试成功!')
```

显示测试失败，原因是：如果我们执行以下命令

```
mul()
```

函数返回1，这不是我们想要的乘积函数的结果，这时候的结果应该是报错，或者提醒我们应该至少输入1个参数。

第四步 改进

```python
def mul(x, *y):
    res = x
    for i in y:
        res = res * i
    return res
```
