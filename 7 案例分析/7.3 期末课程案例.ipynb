## 2. 报告选题参考
在电商平台上分析某品牌手机销售量的影响因素，应该采集哪些相关的数据信息？这些信息如何影响手机的销量？

学生A回答：
（1）对于手机销量的影响因素，应该采集的相关数据有
1. 手机的品牌
1. 手机内使用的技术 
1. 手机的推销力度 
1. 价格 
1. 手机屏幕的大小
1. 手机续航能力
1. 手机的外观
1. 手机发布时间

（2）为何会影响
1. 用户对该品牌手机的喜爱程度，由于品牌效应，很多喜欢该品牌的用户会更有可能成为消费者
1. 手机使用的技术，可以影响手机的质量的好坏和处理速度
1. 用户从何处看到该手机相关信息，投放的广告方式使用户得知改手机的销售，吸引用户前来
1. 用户对价格的心理预期，价格预期很大程度上决定了用户是否会进行购买的可能性
1. 手机屏幕的大小对于有些人来说非常重要，有些人喜欢大屏幕，有些则喜欢小屏幕
1. 用户一般偏向续航能力更强的手机
1. 不同年龄段用户在外观上的审美也不相同
1. 手机的发布时间长短也是一个关键因素，有的人追求新机，有的人选择降价后的发布有段时间的手机

学生B回答：
- 手机销售价格price：通过价格的高低来影响手机的销量
- 手机品牌brand：通过品牌的个性化和独特性等来影响手机的销量
- 手机品牌及型号title：通过对每个型号进行不同性能和配置，外观等来影响手机的销量
- 月销量month_sales：通过判断月销量是否达到某一值，是否会引起跟风现象（羊群效应）来影响手机的销量
- 每款手机电商平台的售卖商家数num：通过商家数来判断手机的市场进而对消费者进行了解，是否进行手机的购买，最终来影响手机的销量。
- 手机主打功能亮点特点tag_info：每个手机亮点不同有的实适合听音乐，有的适合拍照、打游戏等，甚至还有外观的不一致，通过这些不同的主打功能亮点，来影响手机的销量
- 店铺得分 对于手机销量的影响如下：总体而言，手机销量随店铺得分的增高而增高
- 店铺分类 和 累计评论数 对手机销量的影响：官方旗舰店手机销量最高，其它类店铺销量最低。说明人们更倾向于在品质高的店铺购买手机。销量随评论数增高而增高。且离散化的评论数与销量的线性关系更加明显，因此建模时考虑离散化。
- 除此之外：物流、服务、累积评论、CPU、网络类型、屏幕尺寸、手机模式等也是其影响因素的一部分。

而且电商平台上的手机销售数据信息易于爬取，且爬取的结果数据更完整、整洁，极少存在信息的缺漏，并且几乎所有展现出来的信息都是非常规范化的，极大的减少了前期数据预处理（数据清理和数据整理）的工作量。主要是希望通过这个电商平台实际的手机销售数据来分析各类手机品牌目前在市场的一个销售现状。具体来说，主要针对以下几个问题： 
- 手机市场的价格分布； 
- 手机市场的月销量分布； 
- 不同品牌手机总体的月销量数据是怎么样的（从消费者角度出发考虑）； 
- 热卖手机品牌（从售卖商家数的角度出发考虑）； 
- 手机销售市场主打功能卖点亮点分析（词频分析）。

我给出的意见是：
- 顾客：购买手机时，可以优先考虑官方旗舰店，这代表着大多数人的选择。
- 店铺：改善自身服务水平，提升物流速度，进而提升综合评分对于提升销量是不错的选择。
- 制造商：近年来，大屏幕、大内存意味着更高的销量，生产手机时可优先考虑。另外，双卡双待、4G全网通的手机可能更受欢迎。
